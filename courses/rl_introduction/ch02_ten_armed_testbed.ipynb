```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from tqdm import trange



from rl_introduction_course_code.chapter02.ten_armed_testbed import figure_2_1,simulate
from ch02_ten_armed_testbed import Bandit

matplotlib.use('TkAgg')
np.set_printoptions(linewidth=200)
%matplotlib inline
```

```python
runs=2000
steps=1000
epsilon = 0.1

# plt.plot(average_reward)
```

```python
epsilons = [0, 0.1, 0.01]
bandits = [Bandit(epsilon=eps,sample_averages=True) for eps in epsilons]
best_action_counts, rewards = simulate(runs, steps, bandits)
```

```python
# matplotlib.use('TkAgg')
%matplotlib inline
def plot_figure_2_2(rewards,best_action_counts):
    plt.figure(figsize=(10, 20))

    plt.subplot(2, 1, 1)
    for eps, reward in zip(epsilons, rewards):
        plt.plot(reward, label='$\epsilon = %.02f$' % (eps))
        # plt.show()
    plt.xlabel('steps')
    plt.ylabel('average reward')
    plt.legend()
   

    plt.subplot(2, 1, 2)
    for eps, counts in zip(epsilons, best_action_counts):
        plt.plot(counts, label='$\epsilon = %.02f$' % (eps))
    plt.xlabel('steps')
    plt.ylabel('% optimal action')
    plt.legend()
    plt.show()

plot_figure_2_2(rewards,best_action_counts)
```

```python
from rl_introduction_course_code.chapter02.ten_armed_testbed import Bandit as Bandit0
epsilons = [0, 0.1, 0.01]
bandits = [Bandit0(epsilon=eps) for eps in epsilons]
best_action_counts_0, rewards_0 = simulate(runs, steps, bandits)
```

```python
plot_figure_2_2(rewards_0,best_action_counts_0)
```
